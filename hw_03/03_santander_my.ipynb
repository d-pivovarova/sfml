```python
from sklearn.metrics import roc_auc_score, roc_curve
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

penalty = ['l1', 'l2']
C = [0.01, 1.0, 2.5, 10.0, 100.0]

plt.figure(figsize=(10, 10))
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve LogR')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
plt.xticks(np.arange(0.0, 1.0, 0.1))
plt.yticks(np.arange(0.0, 1.0, 0.1))

for p in penalty:
    for c in C:
        lr = LogisticRegression(penalty = p, C = c)
        lr.fit(X_train, y_train)
        lr_predict = lr.predict_proba(X_test)[:,1]
        lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test, lr_predict)
        plt.plot(lr_fpr, lr_tpr, label = 'penalty = ' + str(p) + ', C = ' + str(c))
        plt.legend(loc='lower right')
```

```python
import math
```

```python
def predict(features, coef, last_coef):
    s = 0
    for x1, x2 in zip(features, coef):
        s += x1*x2
    s += last_coef
    return (
        1/(
            1 + math.e**(
                -1 * (
                    s
                )
            )
        )
    )
```
