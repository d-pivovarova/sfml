```python
# Как поступить со строковыми переменными?

# Ваш код :-)
# закодируем значения в новом стобце
obj_cols = []
for i in all_data.columns:
    if str(all_data[i].dtype) == 'object':
        print(i)
        obj_cols.append(i)
#         print(set(all_data[i]))
#         print()
# obj_cols

all_data = pd.get_dummies(all_data, columns = obj_cols)
```

```python
# Вновь разделите на train и test (не мешая)

# Ваш код :-)
# разделим двумя способами: оставив ID объекта исключив атрибут с ID объекта
train_new = all_data[all_data['AGREEMENT_RK'].isin(train['AGREEMENT_RK'])]
train_new_woID = train_new.drop('AGREEMENT_RK', axis=1)
test_new = all_data[all_data['AGREEMENT_RK'].isin(test['AGREEMENT_RK'])]
test_new_woID = test_new.drop('AGREEMENT_RK', axis=1)
print(y.shape, train.shape, train_new.shape, train_new_woID.shape, test.shape, test_new.shape, test_new_woID.shape, all_data.shape)
# all_data = all_data.drop('AGREEMENT_RK', axis=1)
```

```python
# Сделать hold-out sampling на train выборке
from sklearn.cross_validation import train_test_split

# test_size=0.3, random_state=42

# Ваш код :-)
# X_train, X_test, y_train, y_test = train_test_split(train_new, y, test_size=0.3, random_state=42) # лучше не использовать выборку с ID
X_train, X_test, y_train, y_test = train_test_split(train_new_woID, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression


# Ваш код :-)
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import precision_score, recall_score
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

# Ваш код :-)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
predict_dtc = dtc.predict(X_test)
predict_proba_dtc = dtc.predict_proba(X_test)


print('precision', precision_score(y_test, predict_dtc))
print('recall', recall_score(y_test, predict_dtc))
print('accuracy', accuracy_score(y_test, predict_dtc))
print('f1', f1_score(y_test, predict_dtc))
```

```python
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
predict_rfc = rfc.predict(X_test)
predict_proba_rfc = rfc.predict_proba(X_test)

print('precision', precision_score(y_test, predict_rfc))
print('recall', recall_score(y_test, predict_rfc))
print('accuracy', accuracy_score(y_test, predict_rfc))
print('f1', f1_score(y_test, predict_rfc))
```

```python
logr = LogisticRegression(random_state=42)
logr.fit(X_train, y_train)
predict_logr = logr.predict(X_test)
predict_proba_logr = logr.predict_proba(X_test)

print('precision', precision_score(y_test, predict_logr))
print('recall', recall_score(y_test, predict_logr))
print('accuracy', accuracy_score(y_test, predict_logr))
print('f1', f1_score(y_test, predict_logr))
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально посмотреть)

# Ваш код :-)
```

```python
precision_dtc, recall_dtc, thresholds_dtc = precision_recall_curve(y_test, predict_proba_dtc[:,1])
precision_rfc, recall_rfc, thresholds_rfc = precision_recall_curve(y_test, predict_proba_rfc[:,1])
precision_logr, recall_logr, thresholds_logr = precision_recall_curve(y_test, predict_proba_logr[:,1])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(5, 5))
plt.plot(precision_dtc, recall_dtc, label = 'DTC')
plt.plot(precision_rfc, recall_rfc, label = 'RFC')
plt.plot(precision_logr, recall_logr, label = 'LogR')
plt.legend(loc='upper right')
plt.ylabel('recall')
plt.xlabel('precision')
plt.grid(True)
plt.title('PR curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе

# Ваш код :-)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
fpr_dtc, tpr_dtc, thresholds_dtc = roc_curve(y_test, predict_proba_dtc[:,1])
fpr_rfc, tpr_rfc, thresholds_rfc = roc_curve(y_test, predict_proba_rfc[:,1])
fpr_logr, tpr_logr, thresholds_logr = roc_curve(y_test, predict_proba_logr[:,1])
plt.figure(figsize=(5, 5))
plt.plot(fpr_dtc, tpr_dtc, label = 'DTC')
plt.plot(fpr_rfc, tpr_rfc, label = 'RFC')
plt.plot(fpr_logr, tpr_logr, label = 'LogR')
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.legend(loc='lower right')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
roc_auc_score(y_test, predict_proba_dtc[:,1]), roc_auc_score(y_test, predict_proba_rfc[:,1]), roc_auc_score(y_test, predict_proba_logr[:,1])
```

```python
from sklearn.cross_validation import cross_val_score
# Сделать k-fold (10 фолдов) кросс-валидацию каждой модели
# И посчитать средний roc_auc

# Ваш код :-)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
result_dtc = cross_val_score(
    dtc,
    train_new_woID,
    y,
    scoring='roc_auc',
    cv=10
)
print('DTC', result_dtc)

rfc = RandomForestClassifier(random_state=42)
result_rfc = cross_val_score(
    rfc,
    train_new_woID,
    y,
    scoring='roc_auc',
    cv=10
)
print('RFC', result_rfc)

logr = LogisticRegression(random_state=42)
result_logr = cross_val_score(
    logr,
    train_new_woID,
    y,
    scoring='roc_auc',
    cv=10
)
print('LogR', result_logr)
print('-----')
print('DTC', result_dtc.mean())
print('RFC', result_rfc.mean())
print('LogR', result_logr.mean())
```
